## Data Ingestion

```python
data=pd.read_excel("musk_csv.xlsx")
```

```python
data.tail()
```

```python
# separating independent features for later use if needed
```

```python
xx=data.drop("class",axis=1)
yy=data["class"]
```

```python
# Checking the shape of data
```

```python
data.shape
```

```python
# Lets check for Class counts
```

```python
data["class"].value_counts().plot(kind="bar")
```

```python
# As we can clearly see the data is unbalance and thus would need treatment in order to build a model 
```

```python
# Dropping off row specific columns that are not needed
```

```python
data.drop(["molecule_name","ID","conformation_name"],inplace=True,axis=1)
```

```python
data.head()
```

```python
# importing tensorflow, keras and other packages
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.layers import Conv1D, MaxPool1D
from tensorflow.keras.optimizers import Adam
```

```python
# library to split and scale the data
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# now lets first seperate data according to their respective class
```

```python
non_musk = data[data['class']==0]
musk = data[data['class']==1]
```

```python
# checking the shape
```

```python
non_musk.shape, musk.shape
```

```python
# Now the idea here is to take out 1017 rows from the non_musk data through random selection and join them with musk_data to make a 
# balanced dataset
```

```python
non_musk = non_musk.sample(musk.shape[0]) # random selection
non_musk.shape
```

```python
# appending the selected rows
```

```python
data = musk.append(non_musk, ignore_index=True)
data.head()
```

```python
data.shape
```

```python
# Seperating the target and independent features
```

```python
X = data.drop('class', axis = 1)
y = data['class']
```

```python
# performing the train test split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0, stratify = y)    # spliting at 80-20 ratio
```

```python
# checking the shape of training and validation dataset
```

```python
X_train.shape, X_test.shape
```

```python
# now we gonna need to scale our data before building our model
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Converting our scaled dataset into arrays
```

```python
y_train = y_train.to_numpy()
y_test = y_test.to_numpy()
```

```python
X_train.shape
```

## Reshaping the data

```python
# Preparing the data in 3D so that we can pass 2d in iteration 
```

```python
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)     # Adding a third dimension to our data
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)
```

```python
# Verying the shape
X_train.shape, X_test.shape
```

# Building CNN model

```python
# keeping the epochs = 30 for now, will change if required
epochs=30

# Building 1st Layer
model = Sequential()
model.add(Conv1D(32, 2, activation='relu', input_shape = X_train[0].shape))
model.add(BatchNormalization())
model.add(Dropout(0.2))                              # 20% of the neurons will be dropped after 1st layer randomly

# Building 2nd Layer
model.add(Conv1D(64, 2, activation='relu'))     # we removed the input shape as its no longer required
model.add(BatchNormalization())
model.add(Dropout(0.5))                  # 50% of the neurons will be dropped after 2nd layer


# Now we gonna need to convert this multi dimensional data into a vector and thus would be using flattening layer
model.add(Flatten())
# Adding a Dense Layer
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))

# Since its a binary classification we gonna use sigmoid as our activation function
model.add(Dense(1, activation='sigmoid'))   
```

```python
# Lets Check Summary of our model
```

```python
model.summary()
```

```python
# Lets begin to compile our model and for that i am gonna be use Adam optimizer with learning rate = 0.0001
```

```python
model.compile(optimizer=Adam(lr=0.0001), loss = 'binary_crossentropy', metrics=['accuracy'])
```

```python
# Lets now train our model
```

```python
history = model.fit(X_train, y_train, epochs=epochs, validation_data=(X_test, y_test), verbose=1)
```

```python
def plot_learningCurve(history, epoch):
  # Plot training & validation accuracy values
  epoch_range = range(1, epoch+1)
  plt.figure(figsize=[25,10])  
  plt.plot(epoch_range, history.history['accuracy'])
  plt.plot(epoch_range, history.history['val_accuracy'])
  plt.title('Model accuracy')
  plt.ylabel('Accuracy')
  plt.xlabel('Epoch')
  plt.legend(['Train', 'Val'], loc='upper left')
  plt.show()

  # Plot training & validation loss values
  plt.figure(figsize=[25,10])
  plt.plot(epoch_range, history.history['loss'])
  plt.plot(epoch_range, history.history['val_loss'])
  plt.title('Model loss')
  plt.ylabel('Loss')
  plt.xlabel('Epoch')
  plt.legend(['Train', 'Val'], loc='upper left')
  plt.show()
```

```python
plot_learningCurve(history, epochs)
```

```python
c=model.predict_classes(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,c))
```

```python
# Adding MaxPool Layer and increasing the epochs 
```

```python
# keeping the epochs = 100
epochs=100

# Building 1st Layer
model = Sequential()
model.add(Conv1D(32, 2, activation='relu', input_shape = X_train[0].shape))
model.add(BatchNormalization())
model.add(MaxPool1D(2))
model.add(Dropout(0.2))                              # 20% of the neurons will be dropped after 1st layer randomly

# Building 2nd Layer
model.add(Conv1D(64, 2, activation='relu'))     # we removed the input shape as its no longer required
model.add(BatchNormalization())
model.add(MaxPool1D(2))
model.add(Dropout(0.5))                  # 50% of the neurons will be dropped after 2nd layer


# Now we gonna need to convert this multi dimensional data into a vector and thus would be using flattening layer
model.add(Flatten())
# Adding a Dense Layer
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))

# Since its a binary classification we gonna use sigmoid as our activation function
model.add(Dense(1, activation='sigmoid'))   
```

```python
# Lets begin to compile our model and for that i am gonna be use Adam optimizer with learning rate = 0.0001
```

```python
model.compile(optimizer=Adam(lr=0.0001), loss = 'binary_crossentropy', metrics=['accuracy'])
```

```python
# Lets now train our model
```

```python
history = model.fit(X_train, y_train, epochs=epochs, validation_data=(X_test, y_test), verbose=1)
```

```python
plot_learningCurve(history, epochs)
```

```python
c=model.predict_classes(X_test)
```

```python
print(classification_report(y_test,c))
```
